```python
from environments.cleanup_env import CleanupEnv
```

```python
env = CleanupEnv()
```

```python
env.reset()
env.render()
```

```python
run_length = 20
for i in range(run_length):
    agents = list(env.agents.values())
    agent_action_dict = dict()
    for agent in agents:
        action = env.action_space.sample()
        agent_action_dict[agent.agent_id] = action
    obs, reward, dones, truncated, info = env.step(agent_action_dict)
    env.render()
        
```
